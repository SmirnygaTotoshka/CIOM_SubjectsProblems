```python
import pandas as pd
testTab = pd.read_excel(r"C:\Users\SmirnygaTotoshka\Desktop\Problems.xlsx", sheet_name = 1)
testTab['ОУП'] = testTab['ОУП']*100
testTab['ПО'] = testTab['ПО']*100
testTab['КП'] = testTab['КП']*100
testTab['МТО'] = testTab['МТО']*100
testTab['АОС'] = testTab['АОС']*100
```

```python
direction = ['Лечебное дело',"Педиатрия"]
import matplotlib.pyplot as plt
from numpy import arange
for course in range(1,8):
    if (course in list(testTab['Курс'])):
        for d in direction:
            t = testTab[testTab['Специальность'] == d]
            subjects = list(t['Дисциплина'])
            hist_OUP = list(t['ОУП'])
            hist_PO = list(t['ПО'])
            hist_KP = list(t['КП'])
            hist_MTO = list(t['МТО'])
            hist_AOS = list(t['АОС'])

            fig,ax = plt.subplots(figsize=(6, 6),dpi = 120)
            x = arange(len(subjects))
            w = 0.1
            ax.bar(x, hist_OUP, w, color='#CACA86', label='ОУП')
            ax.bar(x + w, hist_PO, w, color='#CAA0BD', label='ПО')
            ax.bar(x + (2 * w),hist_KP, w, color='#4AEDBD', label='КП')
            ax.bar(x + (3 * w), hist_MTO, w, color='#FBABF5', label='МТО')
            ax.bar(x + (4 * w), hist_AOS, w, color='#FEB35E', label='АОС')

            ax.set_ylabel('Процент учащихся')
            ax.set_ylim(0,100)
            ax.set_xticks(x + w + w/2)
            ax.set_xticklabels(subjects)
            ax.set_xlabel('Предмет')
            ax.set_title(d + " " + str(course) + " курс")
            ax.legend()
            plt.grid(True, 'major', 'y', ls='--', lw=.5, c='k', alpha=.3)

            fig.tight_layout()
            plt.show()
```
