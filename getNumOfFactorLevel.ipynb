```python
#Для ответов с множественным выбором
#Подсчитывает количество ответов для каждого варианта
#Возвращает словарь: ключ - вариант ответа, значение - количество ответов
#Параметры:
#pd - объект библиотеки pandas(import pandas as pd), Именно этот, не дата фрейм или что-то другое
#factor - Столбец-фактор, содержащий ответы
#delimiter - Символ разделяющий варианты ответов в ячейке(не уверен, можно ли regexp)
#author - SmirnygaTotoshka
def getNumOfEachFactorLevel(pd,factor,delimiter = ','):
    result = {"NA":0}
    for value in factor:
        if (pd.isna(value)):
            result["NA"] += 1
        else:
            answer = value.split(delimiter)
            for a in answer:
                key = a.strip()
                if (not key in result.keys()):
                    result[key] = 1
                else:
                    result[key] += 1
    return result

                
```

```python
res = getNumOfEachFactorLevel(pd,testTab['С какими проблемами вы сталкивались? [Анатомия]'])
res
```
